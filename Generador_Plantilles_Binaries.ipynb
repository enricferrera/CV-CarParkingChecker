```python
import cv2
import numpy as np
import os

output_dir = "plantilles"
os.makedirs(output_dir, exist_ok=True)
os.makedirs(f"{output_dir}/numeros", exist_ok=True)
os.makedirs(f"{output_dir}/lletres", exist_ok=True)

font = cv2.FONT_HERSHEY_DUPLEX  
font_scale = 2.5
thickness = 7
size = (40, 60)

for char in "0123456789":
    img = np.ones((60, 40), dtype=np.uint8) * 255  # blanc de fons
    text_size = cv2.getTextSize(char, font, font_scale, thickness)[0]
    text_x = (img.shape[1] - text_size[0]) // 2
    text_y = (img.shape[0] + text_size[1]) // 2
    cv2.putText(img, char, (text_x, text_y), font, font_scale, 0, thickness, cv2.LINE_AA)
    cv2.imwrite(f"{output_dir}/numeros/{char}.jpg", img)

for char in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
    img = np.ones((60, 40), dtype=np.uint8) * 255  # blanc de fons
    text_size = cv2.getTextSize(char, font, font_scale, thickness)[0]
    text_x = (img.shape[1] - text_size[0]) // 2
    text_y = (img.shape[0] + text_size[1]) // 2
    cv2.putText(img, char, (text_x, text_y), font, font_scale, 0, thickness, cv2.LINE_AA)
    cv2.imwrite(f"{output_dir}/lletres/{char}.jpg", img)
```
